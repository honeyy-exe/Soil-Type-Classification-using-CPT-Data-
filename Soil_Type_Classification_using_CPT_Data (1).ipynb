Importing the dependencies


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

Data Collection and Processing

```python
# loading the dataset to pandas dataframe
dataset = pd.read_csv("/WW10.csv")
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
# to find the number of missing values in each column to later clean those rows
dataset.isnull().sum()
```

```python
# So no row in the cell is empty and we can proceed with the given data
dataset['Rf'] = ((dataset['sleeve'] / 1000)/dataset['tip'])*100
```

```python
# checking if the new column ' Rf' is created
dataset.head()
```

```python
# Classifying the soil type based on the Rf value we derived, function of determining the soil type based on the Rf value
def classify_soil(Rf):
  if Rf <= 1:
    return 'Sand'
  elif Rf <= 4:
    return 'Silty Sand'
  elif Rf <= 7:
    return 'Silt'
  else:
    return 'Clay'

#Adding another column into the dataset after applying the function 'classify_soil' to the Rf value in the data
dataset['soil_type'] = dataset['Rf'].apply(classify_soil)

#checking the data if the row 'soil_type' got added
dataset.head()
```

Machine Learning Model Development

```python
#Giving the Machine Learning Model its inputs and target variable
X = dataset[['tip' , 'sleeve' , 'z']] #inputs
Y = dataset[['soil_type']] #target variable
```

```python
#Splitting the data into target data and test data
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2, random_state = 2 )
```

```python
#Bringing random forest model into action
model = RandomForestClassifier()
# Assigning the train data to the model to understand and evaluate
model.fit(X_train , Y_train)
```

```python
#Testing the model with the test data we seperated from our base data
prediction = model.predict(X_test)
```

Model Evaluation

```python
#Now the model is evaluated and the accuracy is calculated
print('Accuracy : ' , accuracy_score(Y_test , prediction))
```

```python
# now we need the classification report
print(classification_report(Y_test , prediction))
```
